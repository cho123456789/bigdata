```python
# 데이터 유형 파악 

import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_test.csv")


# display(x_train.head())
# display(y_train.head())

# print(x_train.info())
# print(x_test.info())
# print(y_train)

# 결측값 확인
# print(x_train.isnull().sum())
# print(x_test.isnull().sum())
# print(y_train.isnull().sum())

# 결측값 대채
x_train["Arrival Delay in Minutes"].fillna(x_train["Arrival Delay in Minutes"].mean())
x_test["Arrival Delay in Minutes"].fillna(x_train["Arrival Delay in Minutes"].mean())

# 데이터 전처리 
X_train = x_train.drop(["ID","id"], axis= 1)
X_test = x_test.drop(["ID","id"], axis= 1)
Y_train = y_train["satisfaction"]

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

# 테스트 데이터 와 학습데이터 컬럼 수 맞춤
X_test = X_test.reindex(columns=X_train.columns, fill_value = 0)

# 데이터 분리 
from sklearn.model_selection import train_test_split
X_train, X_val , Y_train , Y_val = train_test_split(X_train, Y_train, test_size = 0.2)

# print(X_train.shape, X_val.shape , Y_val.shape, Y_train.shape)

# 데이터 학습 

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,Y_train)
#y_pred_val = model.predict(X_val)
# print(y_pred_val)

# 데이터 평가 
from sklearn.metrics import roc_auc_score , accuracy_score

predict_train_label = model.predict(X_train)
predict_train_proba = model.predict_proba(X_train)[:,1]
predict_validation_label = model.predict(X_val)
predict_validation_prob = model.predict_proba(X_val)[:,1]

print('validation auc :', roc_auc_score(Y_val,predict_validation_prob))

# print(predict_train_proba)
# model.classes_

predict_test_prob = model.predict_proba(X_test)[:,0]
print(predict_test_prob)

# 데이터 저장 
result = pd.DataFrame(predict_test_prob, columns = ["pred"])
result.to_csv("result.csv",index=False)

# 데이터 읽기
result = pd.read_csv("result.csv")
result.head()
```

```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/x_test.csv")


# display(x_train.head())
# display(y_train.head())

# print(x_train.info())
# print(x_test.info())

# 데이터 정제 
X_train = x_train.drop(["ID"],axis = 1)
X_test = x_test.drop(["ID"],axis = 1)
Y_train = y_train["Drug"]

# 결측치 제거 
# X_train.isnull().sum()
# X_test.isnull().sum()
 
#  데이터 인코딩 
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

X_test = X_test.reindex(columns = X_train.columns , fill_value = 0)


# 데이터 분리 
from sklearn.model_selection import train_test_split

X_train , X_val , Y_train, Y_val = train_test_split(X_train, Y_train, test_size = 0.2)

# 데이터 학습 
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, Y_train)
y_pre_val = model.predict(X_val)
y_pre_pro = model.predict_proba(X_val)
print(y_pre_val)

# 데이터 평가 
from sklearn.metrics import roc_auc_score , accuracy_score
roc_auc = roc_auc_score(Y_val, y_pre_pro , multi_class ="ovr")
acc_score = accuracy_score(Y_val, y_pre_val)
print(roc_auc)
print(acc_score)

# 데이터 저장 
y_pred = model.predict(X_test)
result = pd.DataFrame(y_pred , columns = ["pred"])
result.to_csv("result.csv",index= False)

# 데이터 읽기 
result = pd.read_csv("result.csv")
print(result)
```
