```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv')

display(train.head(2))
test.head(2)

# train.isnull().sum()
# test.isnull().sum()

X_train = train.drop(["ID","이용금액"] , axis =1)
X_test = test.drop(["ID"], axis = 1)
Y_train = train["이용금액"]


X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
X_test = X_test.reindex(columns = X_train.columns , fill_value = 0)


from sklearn.model_selection import train_test_split

X_train, X_val , Y_train, Y_val = train_test_split(X_train, Y_train, test_size = 0.2)

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, Y_train)
y_pred_val = model.predict(X_val)
# print(y_pred_val)

from sklearn.metrics import mean_squared_error
import numpy as np
rsme =  np.sqrt(mean_squared_error(y_pred_val , Y_val))
print(rsme)

y_pred = model.predict(X_test)
result = pd.DataFrame(y_pred, columns = ["pred"])
result = pd.concat([test["ID"],result] , axis =1 )
result.to_csv('result.csv',index= False)

result = pd.read_csv('result.csv')
print(result)
```

```python
import pandas as pd 
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p3_t.csv')
df.head()
```

```python
# train 데이터로 target을 종속변수로 로지스틱 회귀를 진행할 때 age 컬럼의 오즈비를 구하여라
```

```python
from statsmodels.formula.api import logit
import numpy as np

train_data = df.iloc[:210, :].copy()  # 앞 210개 행
test_data = df.iloc[210:, :].copy()   #

formula = ('target ~ age+sex+cp+trestbps+chol+fbs+restecg+thalach+exang+oldpeak+slope+ca+thal')
model = logit(formula ,data =train_data).fit()
# print(model.summary())

print(np.exp(model.params['age']))
residual_deviance = -2 * model.llf  # llf: 최대화된 로그우도(log-likelihood)
print("Residual Deviance:", residual_deviance)
```

```python
# train으로 로지스틱 회귀 진행했을 경우 로짓 우도값을 도출하라
residual_deviance = model.llf
print("Residual Deviance:", residual_deviance)
```

```python
# test 데이터의 독립변수로 target 예측 후 오류율을 구하여라  # train 데이터는 앞의 210개 행을, test데이터는 나머지 부분을 사용한다
```

```python
from  sklearn.metrics import accuracy_score 
train_pred = model.predict(test_data)

test_pred = (train_pred > 0.5).astype(int)

acc = accuracy_score(test_data['target'], test_pred)
print("Accuracy:", acc)

error_rate = 1 - acc
print("Error Rate:", error_rate)
```
