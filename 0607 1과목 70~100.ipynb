```python
# 전체데이터의 수축기혈압(최고) - 이완기혈압(최저)의 평균을 구하여라
```

```python
print((df['수축기혈압(최고) : mmHg']- df['이완기혈압(최저) : mmHg']).mean())
```

```python
# 50~59세의 신장평균을 구하여라
```

```python
df['나이대'] = (df['측정나이'] // 10) * 10
df[df['나이대'] == 50]['신장 : cm'].mean()
```

```python
df['나이대'].value_counts()
```

```python
# 연령대 (20~29 : 20대 …) 별 등급의 숫자를 데이터 프레임으로 표현하라
df.groupby(['나이대','등급']).size().to_frame(name='size').reset_index()
```

```python
# 남성 중 A등급과 D등급의 체지방률 평균의 차이(큰 값에서 작은 값의 차)를 구하여라
```

```python
m = df[(df['측정회원성별'] == 'M') & (df['등급'] == 'A')]['체지방율 : %'].mean()
f = df[(df['측정회원성별'] == 'M') & (df['등급'] == 'D')]['체지방율 : %'].mean()
print(f-m)
```

```python
#여성 중 A등급과 D등급의 체중의 평균의 차이(큰 값에서 작은 값의 차)를 구하여라
m = df[(df['측정회원성별'] == 'F') & (df['등급'] == 'A')]['체중 : kg'].mean()
f = df[(df['측정회원성별'] == 'F') & (df['등급'] == 'D')]['체중 : kg'].mean()
print(f-m)
```

```python
#bmi는 자신의 몸무게(kg)를 키의 제곱(m)으로 나눈값이다. 데이터의 bmi 를 구한 새로운 컬럼을 만들고 남성의 bmi 평균을 구하여라
```

```python
df['bmi'] = df['체중 : kg'] / (df['신장 : cm'] / 100)**2
df[df['측정회원성별'] == 'M']['bmi'].mean()
```

```python
# bmi보다 체지방율이 높은 사람들의 체중평균을 구하여라
```

```python
df[(df['bmi'] <= df['체지방율 : %'])]['체중 : kg'].mean()
```

```python
# 남성과 여성의 악력 평균의 차이를 구하여라
```

```python
df
m = df[df['측정회원성별'] == 'M']['악력D : kg'].mean()
f = df[df['측정회원성별'] == 'F']['악력D : kg'].mean()
print(m-f)
```

```python
# 남성과 여성의 교차윗몸일으키기 횟수의 평균의 차이를 구하여라
```

```python
m = df[df['측정회원성별'] == 'M']['교차윗몸일으키기 : 회'].mean()
f = df[df['측정회원성별'] == 'F']['교차윗몸일으키기 : 회'].mean()
print(m-f)
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/mobile/train.csv")
df.head()
```

```python
# price_range 의 각 value를 그룹핑하여 각 그룹의 n_cores 의 빈도가 가장높은 value와 그 빈도수를 구하여라
```

```python
result = (
    df[['price_range', 'n_cores']]
    .groupby(['price_range', 'n_cores'])
    .size()
    .reset_index(name='count')
)
result = result.sort_values(['price_range', 'count'], ascending=[True, False])
result = result.groupby('price_range').first().reset_index()
result
```

```python
result= df[df['price_range'] == 3].corr().unstack()
result = result[result!=1]
result.sort_values(ascending = False) # pc           fc              0.635166
```

```python
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv")
train.head()
```

```python
# 흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 평균의 차이는?
```

```python
results = train[train['smoker'] == 'yes']['charges']
resultn = train[train['smoker'] == 'no']['charges']
results = results[results >= results.quantile(0.9)].mean()
resultn = resultn[resultn >= resultn.quantile(0.9)].mean()

print(results - resultn)
```
