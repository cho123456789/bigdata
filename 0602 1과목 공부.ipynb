```python
# **5대 범죄(절도, 사기, 배임, 방화, 폭행)의 월별 총 발생건수를 총범죄수라고 표현하자. 
# 18,19년의 각각 분기별 총범죄수의 월평균 값을 구했을때 최대값을 가지는 년도와 분기를 구하고 
# 해당 분기의 최댓값의 사기가 발생한 월의 사기 발생건수를 출력하라
# (1분기:1,2,3월 / 2분기 : 4,5,6월 / 3분기 7,8,9월 / 4분기 10,11,12월 , 1분기 월평균 : 1,2,3월의 총범죄수 평균) **
```

```python
df['총범죄수'] = df['절도'] + df['사기'] + df['배임'] + df['방화'] + df['폭행']
df['범죄년도'] = df['범죄분류'].str.replace('_',' ').str.split(' ').str[0]
df['범죄월'] = df['범죄분류'].str.replace('_',' ').str.split(' ').str[1]
def g(x):
    if x == '1월' or x == '2월' or x == '3월':
        return '1분기'
    if x == '4월' or x == '5월' or x == '6월':
        return '2분기'
    if x== '7월' or  x =='8월'  or x == '9월':
        return '3분기'
    else:
        return '4분기'

df['분기'] = df['범죄월'].apply(g)
df.groupby(['분기','범죄년도'])['총범죄수'].mean().sort_values(ascending = False).index[0] # 2019 2분기

result = df[df['분기'] =='2분기']['사기'].max() # 27766
df[df['사기'] == result]['사기'].sum()
```

```python
# BMI는 몸무게(kg) / (키(M) * 키(M)) 로 정의 된다. 초고도 비만은 BMI 25이상 , 
# 고도 비반은 BMI 25미만 - 23이상 , 
# 정상은 23미만 - 18.5이상 
# 저체중은 18.5미만으로 정의 된다. 
#주어진 데이터에서 초고도비만 인원 + 저체중 인원 의 숫자는?
```

```python
df['BMI'] = df['weight(kg)'] / (df['height(cm)'] / 100) ** 2

def g(x):
    if x > 25:
        return '초고도 비만'
    if x < 25 and x >= 23:
        return '고도비만'
    if x < 23 and x >= 18.5:
        return '정상'
    if x < 18.5:
        return '저체중'

df['비만도']  = df['BMI'].apply(g)

df['비만도'].value_counts()  

8978+ 20
```
