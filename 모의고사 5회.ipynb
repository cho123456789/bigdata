```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_2_.csv')
df.head(5)
```

```python
# BMI는 몸무게(kg) / (키(M) * 키(M)) 로 정의 된다. 
#초고도 비만은 BMI 25이상 , 고도 비반은 BMI 25미만 - 23이상 , 정상은 23미만 - 18.5이상
#저체중은 18.5미만으로 정의 된다. 주어진 데이터에서 초고도비만 인원 + 저체중 인원 의 숫자는?
```

```python
df['BMI'] = df['weight(kg)']  / (df['height(cm)']/100) ** 2

def g(x):
    if x >= 25:
        return '초고도비만'
    if x < 25 and x >= 23:
        return '고도 비만'
    if x < 23 and x >= 18.5:
        return '정상'
    if x < 18.5:
        return '저체중'

df['비만도'] = df['BMI'].apply(g)
len(df[df['비만도'] == '초고도비만']) + len(df[df['비만도'] == '저체중'])
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_3.csv')
df.head(5)
```

```python
# 순유입인원은 초중고 도내,도외 전입인원에서 초중고 도내, 도외 전출인원을 뺀값이다. 
#각년도별로 가장 큰 순유입인원을 가진 지역구의 순유입인원을 구하고 
#전체 기간의 해당 순유입인원들의 합을 구하여라
```

```python
df['전출도내'] = df['초등학교_전출_도내'] + df['중학교_전출_도내'] + df['고등학교_전출_도내']
df['전출도외'] = df['초등학교_전출_도외'] + df['중학교_전출_도외'] + df['고등학교_전출_도외']

df['전입도내'] = df['초등학교_전입_도내'] + df['중학교_전입_도내'] + df['고등학교_전입_도내']
df['전입도외'] = df['초등학교_전입_도외'] + df['중학교_전입_도외'] + df['고등학교_전입_도외']
```

```python
df['순유입인원'] = (df['전입도내']+ df['전입도외']) - (df['전출도내']+ df['전출도외'])
result = df.groupby("년도")['순유입인원'].max().sum()
result
```

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_test_.csv')

# display(train.head(2))
# test.head(2)

# train.isnull().sum()
# test.isnull().sum()

X_train = train.drop(["ID","price"], axis = 1)
X_test = test.drop(["ID"], axis = 1)
Y_train = train["price"]

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
X_test = X_test.reindex(columns = X_train.columns, fill_value = 0)

from sklearn.model_selection import train_test_split

X_train, X_val , Y_train, Y_val = train_test_split(X_train,Y_train,test_size= 0.2)

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, Y_train)
y_pre_val = model.predict(X_val)
# print(y_pre_val)

from sklearn.metrics import mean_squared_error
import numpy as np
rmsed = np.sqrt(mean_squared_error(Y_val, y_pre_val))
print(rmsed)

y_pred= model.predict(X_test)
result = pd.DataFrame(y_pred, columns = ["pred"])
result = pd.concat([test["ID"],result], axis=1)
result.to_csv("result.csv", index= False)

result = pd.read_csv("result.csv")
result
```

```python
import pandas as pd 
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p3_1.csv')
df.head()
```

```python
# 이 학생들의 키의 95% 신뢰구간을 구하고자 한다.
```

```python
# 55명 학생들의 키에 대한 표본 평균을 구하여라(반올림하여 소숫점 3째자리까지
```

```python
import numpy as np

mean = np.round(np.mean(df['height']),3)
mean
```

```python
# t분포 양쪽 꼬리에서의 t 값을 구하여라 (반올림하여 소수4째자리까지)
```

```python
import numpy as np
from scipy.stats import t 

std = np.std(df['height'], ddof = 1)
n = len(df['height'])

# 신뢰수준
confidence_level = 0.95
ddof = n - 1
# 자유도 


# t 의 양쪽 꼬리 

t_value = round(t.ppf((1 + confidence_level) / 2, ddof),4)
print(t_value)
```

```python
# 95% 신뢰구간을 구하여라(print(lower,upper) 방식으로 출력, 각각의 값은 소숫점 이하 3째자리까지)
```

```python
lower = round(mean - t_value * std / np.sqrt(n),3)
upper = round(mean + t_value * std / np.sqrt(n),3)
print(lower,upper)
```
