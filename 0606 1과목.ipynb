```python
import pandas as pd


channel =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
video =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
display(channel.head())
display(video.head())
```

```python
# 각 에피소드의 시작날짜(년-월-일)를 에피소드 이름과 묶어 데이터 프레임으로 만들고 출력하라
```

```python
video =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
video['ct'] = pd.to_datetime(video['ct'])
result = video.sort_values(['ct','videoname'], ascending = True).drop_duplicates('videoname')
result['ct'] = result['ct'].dt.date   # date -> 년 월일 
result[['ct', 'videoname']]
```

```python
# “공범” 컨텐츠의 경우 19:00시에 공개 되는것으로 알려져있다. 
#공개된 날의 21시의 viewcnt, ct, videoname 으로 구성된 데이터 프레임을 viewcnt를 내림차순으로 정렬하여 출력하라
```

```python
video['ct'] = pd.to_datetime(video['ct'])
video['hour'] = video['ct'].dt.hour
result = video[video['hour'] == 21]
result = result.sort_values(['videoname','ct']).drop_duplicates('videoname')
result[['viewcnt','ct','videoname']].sort_values('viewcnt',ascending = False)
```

```python
video['time']= video.ct.dt.hour

answer = video.loc[video['time'] ==21] \
            .sort_values(['videoname','ct'])\
            .drop_duplicates('videoname') \
            .sort_values('viewcnt',ascending=False)[['videoname','viewcnt','ct']]\
            .reset_index(drop=True)

display(answer)
```

```python
# video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고
#videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라
```

```python
result = video.sort_values(['videoname','ct']).drop_duplicates('videoname', keep = 'last')
result['ratio'] = result['dislikecnt']/ result['likecnt']
result[['videoname','ratio']].sort_values('ratio')
```

```python
# 2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오
```

```python
start =  pd.to_datetime("2021-11-01 00:00:00")
end = pd.to_datetime("2021-11-01 15:00:00")

target = video[(video['ct'] >= start) & (video['ct'] <= end)]

def g(x):
    result = max(x) - min(x)
    return result

answer = target[['videoname','viewcnt']].groupby('videoname').agg(g)  #agg 집계 합 연산 같은 함수에서 사용 
answer
```

```python
import pandas as pd
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bicycle/seoul_bi.csv')
df.head()
```

```python
# 대여일자별 데이터의 수를 데이터프레임으로 출력하고, 가장 많은 데이터가 있는 날짜를 출력하라
```

```python
df.groupby('대여일자').size().to_frame(name='대여일자')
```

```python
# 각 일자의 요일을 표기하고 (‘Monday’ ~’Sunday’) ‘day_name’컬럼을 추가하고 이를 이용하여 각 요일별 이용 횟수의 총합을 데이터 프레임으로 출력하라
```

```python
df['대여일자'] = pd.to_datetime(df['대여일자'])
df['dayname'] = df['대여일자'].dt.day_name()
df.groupby('dayname').size().sort_values(ascending = False).to_frame(name ='count')
```

```python
# 각 요일별 가장 많이 이용한 대여소의 이용횟수와 대여소 번호를 데이터 프레임으로 출력하라
```

```python
result = df.groupby(['dayname','대여소번호']).size().to_frame('count').sort_values(['dayname','count'], ascending = False).reset_index()
result.drop_duplicates('dayname')
```

```python
# 나이대별 대여구분 코드의 (일일권/전체횟수) 비율을 구한 후 가장 높은 비율을 가지는 나이대를 확인하라. 일일권의 경우 일일권 과 일일권(비회원)을 모두 포함하라
```

```python
df
daily = df[df['대여구분코드'].isin(['일일권','일일권비회원'])].value_counts('연령대코드').sort_values()
total = df.value_counts('연령대코드').sort_values()
total

result = daily / total
result.to_frame('count').sort_values('count',ascending =False)
```

```python
# 연령대별 평균 이동거리를 구하여라
```

```python
df.groupby('연령대코드')['이동거리'].mean()
```

```python
# 연령대 코드가 20대인 데이터를 추출하고,이동거리값이 추출한 데이터의 이동거리값의 평균 이상인 데이터를 추출한다.
#최종 추출된 데이터를 대여일자, 대여소 번호 순서로 내림차순 정렬 후
#1행부터 200행까지의 탄소량의 평균을 소숫점 3째 자리까지 구하여라
```

```python
result = df[df['연령대코드'].str.contains('20대')].reset_index(drop = True)
result = result[result['이동거리'] >= result['이동거리'].mean()].reset_index(drop = True)
result['탄소량'] = result['탄소량'].astype('float')
result = result.sort_values(['대여일자','대여소번호'], ascending= False).reset_index(drop = True).head(200)
round(result['탄소량'].mean(),3)
```

```python
tw = df[df.연령대코드 =='20대'].reset_index(drop=True)
tw_mean = tw[tw.이동거리 >= tw.이동거리.mean()].reset_index(drop=True)
tw_mean['탄소량'] =tw_mean['탄소량'].astype('float')
target =tw_mean.sort_values(['대여일자','대여소번호'], ascending=False).reset_index(drop=True).iloc[:200].탄소량
result = round(target.sum()/len(target),3)
print(result)
```
