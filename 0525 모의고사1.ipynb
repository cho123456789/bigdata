```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
df.head()
```

```python
# 마케팅 응답 고객들의 나이를 10살 단위로 변환 했을 때, 
#가장 많은 인원을 가진 나이대는? (0~9 : 0 , 10~19 : 10)
```

```python
def g(x):
    if x >= 0 and x <= 9:
        return '0대'
    if x >= 10 and x < 20:
        return '10대'
    if x >= 20 and x < 30:
        return '20대'
    if x >= 30 and x < 40:
        return '30대'
    if x >= 40 and x < 50:
        return '40대'
    if x >= 40 and x < 50:
        return '40대'     
    if x >= 50 and x < 60:
        return '50대' 
    if x >= 60 and x < 70:
        return '60대' 
    if x >= 70 and x < 80:
        return '70대' 
    if x >= 80 and x < 90:
        return '80대'
    else:
        return '90대'
df['나이대'] = df['age'].apply(g)
```

```python
import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/test.csv')
submission= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/submission.csv')

# display(train.head())
# display(test.head())
# display(submission.head())
# 은행의 전화 마케팅에 대해 고객의 반응 여부

X_train = train.drop(["ID","y"], axis = 1)
X_test = test.drop(["ID"], axis = 1)
Y_train = train['y']

# X_train.isnull().sum()
# X_test.isnull().sum()
# Y_train.isnull().sum()

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

X_test = X_test.reindex(columns = X_train.columns , fill_value = 0)

# 데이터 분리 

from sklearn.model_selection import train_test_split
X_train, X_val , Y_train, Y_val = train_test_split(X_train, Y_train, test_size = 0.2)
# print(X_train.shape, X_val.shape, Y_train.shape, Y_val.shape)

# 데이터 학습 
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, Y_train)
y_pre_val = model.predict_proba(X_val)
print(y_pre_val)

# 데이터 평가 
from sklearn.metrics import roc_auc_score , accuracy_score
roc_auc = roc_auc_score(Y_val,y_pre_val[:,1])
print(roc_auc)

# 데이터 저장 
y_pred = model.predict_proba(X_test)
result = pd.DataFrame(y_pred[:,1], columns = ["pred"])
result.to_csv("result.csv",index= False)


# 데이터 읽기 
result = pd.read_csv("result.csv")
print(result)
```
