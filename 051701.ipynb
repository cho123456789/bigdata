```python
# import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/train.csv")
df.head()
```

```python
# 98 남성들의 연령대별 (10살씩 구분 0~9세 10~19세 …) 
#Na_to_K값의 평균값을 구해서 데이터 프레임으로 표현하여라
```

```python
df = df[df['Sex'] == 'M']
```

```python
df['나이대'] 
```

```python
df['Age'].value_counts()
```

```python
def g(x):
    if x >= 10 and x < 20:
        return '10대'
    if x >= 20 and x < 30:
        return '20대'
    if x >= 30 and x < 40:
        return '30대'  
    if x >= 40 and x < 50:
        return '40대' 
    if x >= 50 and x < 60:
        return '50대'    
    if x >= 60 and x < 70:
        return '60대'
    else:
        return '70대'

df['나이대'] = df['Age'].apply(g)
```

```python
df.groupby('나이대')['Na_to_K'].mean()
```
