```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e8_p1_3.csv')
df.head(5)
```

```python
# CO(GT),NMHC(GT) 두 컬럼을 min-max normalize를 진행하고 각 컬럼의 모표준편차의 차이의 절댓값을 소숫점 이하 3째 자리까지 구하여라
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(df[['CO(GT)', 'NMHC(GT)']])
df_scaled = pd.DataFrame(scaled_values, columns=['CO(GT)', 'NMHC(GT)'])
df_scaled

std_co = np.std(df_scaled['CO(GT)'], ddof=0)
std_nmhc = np.std(df_scaled['NMHC(GT)'], ddof=0)

print(round(std_co - std_nmhc,3))
```

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e8_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e8_p2_test2.csv')

# display(train.head(2))
# test.head(2)

# train.isnull().sum()
# test.isnull().sum()

X_train = train.drop(["ID","Y"], axis= 1)
X_test = test.drop(["ID"], axis =1)
Y_train = train["Y"]

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
X_test = X_test.reindex(columns = X_train.columns , fill_value = 0)


from sklearn.model_selection import train_test_split
X_train , X_val , Y_train , Y_val = train_test_split(X_train, Y_train, test_size = 0.2)

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, Y_train)
y_pred_val = model.predict(X_val)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_val, y_pred_val)
print(mse)

y_pred = model.predict(X_test)
result = pd.DataFrame(y_pred, columns = ["pred"])
result.to_csv("result.csv", index =False)

result = pd.read_csv("result.csv")
result
```

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e8_p3_1.csv')
df.head()
```

```python
# y컬럼을 종속변수로하는 로지스틱회귀 모형을 만들 떄 유의미하지 않은 변수의 개수는? % 상수항은 추가하여 모델링하라
```

```python
from statsmodels.formula.api import logit
formula="Y~X1+X2+X3+X4+X5+X6+X7+X8+X9+X10+X11+X12+X13+X14+X15"
model = logit(formula, data =df).fit()
print(model.summary()) 

# 유의미 하지않다 => 0.05 보다 크다 # 12개
```

```python
# p-value가 작으면 (예: < 0.05),
# → 귀무가설 기각
# → 변수는 종속변수에 유의미한 영향이 있다.

# p-value가 크면 (≥ 0.05),
# → 귀무가설 기각 못함
# → 변수는 종속변수에 유의미한 영향이 없다.

# 유의미 하다 => 0.05 보다 작다 
# 유의한 변수들 만으로 다시 모델링을 해서 X를 포함하는 독립변수들의 회귀계수 합계를 소숫점 이하 3째 자리까지 구하여라 **
```

```python
from statsmodels.formula.api import logit
formula="Y~X7+X8+X11"
model = logit(formula, data =df).fit()
print(model.summary()) 
```

```python
# 3-2번의 모델 결과에 대해 X뒤의 숫자가 가장 낮은 컬럼이 5단위 늘어나면 오즈비 몇 배로 변화?
```

```python
import pandas as pd 
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e8_p3_2.csv')
df.head()
```

```python
# PIQ를 예측하는 다중선형회귀 모형을 만들고 Brain의 회귀 계수를 구하여라
```

```python
from statsmodels.formula.api import ols
formula = "PIQ ~+Brain+Height+Weight"
model = ols(formula, data = df).fit()
print(model.summary())
print(model.params["Brain"])
print(model.rsquared)
```

```python
# 위에서 계산한 모델을 이용하여 뇌크기 90 키 70 몸무게 150일 때 지능을 예측하라
result = pd.DataFrame({ 'Brain' :[90] , 'Height':[70] , 'Weight' : [150] })
model.predict(result)
```
